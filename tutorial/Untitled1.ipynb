```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def modelA(time, omega):
    return np.sin(omega * time)
```

```python
sigma = 0.5
omega_true = 1.2
time = np.linspace(0, 12, 1000)
ydet = modelA(time, omega=omega_true)
noise = np.random.normal(0, sigma, len(time))
yobs = ydet + noise # Observed data = SIGNAL + GAUSSIAN NOISE

plt.plot(time, yobs, "o")
plt.ylabel("$y_{obs}$"); plt.xlabel("time")
plt.show()
```

```python
plt.plot(time, noise, "o")
```

```python
test1 = (1/np.sqrt(2*np.pi))*(1/sigma)*np.exp(-(noise**2)/(2*sigma**2))
```

```python
# select a point in time, say time[500]
#let there be a distribution with mean ydet[500] and sigma 0.1
# this sigma doesn’t need to match with sigma of noise
hh = np.random.normal(ydet[500], 0.1, 1000)
```

```python
plt.plot(noise, test1, "o")
```

```python
# omega_true = 1.2
sigma2 = 0.1
test2 = (1/np.sqrt(2*np.pi))*(1/sigma2)*np.exp(-((yobs-ydet)**2)/(2*sigma2**2))
plt.plot(noise, test2, "o")
```

```python
result2 = np.sum(np.log(test2))
print(result2)
```
