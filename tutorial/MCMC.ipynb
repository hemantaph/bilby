Metropolis Algorithm

```
#library
from scipy import random
import numpy as np
import matplotlib.pyplot as plt

```

```
a=0
b=np.pi
N=5000

areas = []

for j in range(N):
  xrand=random.uniform(a,b,N)
  integral=0.0

  for i in range(N):
    integral += np.sin(xrand[i])

  ans = (b-a)/float(N)*integral
  areas.append(ans)
 
```

```
plt.title('diatribution of area calculated')
plt.hist(areas, bins=40, density=True, ec='black')
plt.xlabel('areas')
plt.show()
```

```
from scipy.stats import norm
from scipy.stats import mode
'''
areas=np.array(areas)/np.max(areas)'''
mean,std=norm.fit(areas)

plt.hist(areas, bins=30, density=True, ec='black')
xmin, xmax = plt.xlim()

x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mean, std)
plt.plot(x, y)
plt.show()

```
